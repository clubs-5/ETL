# input data:Dataframe.csv 為夥伴整理好資料

# Question:獎項包括人或者影集本身，兩者處理方式不同

# goal:將資料做區分

# output: Emmy_Prime_only_film.csv

```python
 Emmy_data = pd.read_csv('./Clean_DataFrame.csv')
 Emmy_data
```

```python
Emmy_data.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
Emmy_data
```

## 將欄位名稱重新命名

```python
Emmy_data.rename(columns={"Finalist":"Title","producer":'Director',"producer_2":'Director_2'},inplace=True)
```

```python
Emmy_data.drop(['Director', 'Director_2'], axis=1,inplace=True)
```

```python
Emmy_data['Director'] = Emmy_data.Groups.str.split(',').str.get(0)
```

```python
Emmy_data['Director_2'] = Emmy_data.Groups.str.split(',').str.get(1)
```

#  區分方式: 觀察資料發現，Director_2為NA值資料為獎項為個人

```python
Emmy_data.Director_2.isna().sum()
```

```python
 #Emmy_data_film is only film
 Emmy_data_film = Emmy_data.dropna(subset=['Director_2'])
```

# 由於該年該獎項沒有重複，對年分以及得獎影集做groupby 總和

```python
Emmy_data_film = Emmy_data_film.groupby(['Years','Title']).sum()
```

```python
Emmy_data_film = Emmy_data_film.reset_index()
```

```python
Emmy_data_film[Emmy_data_film['Title'] =='Succession']
```

```python
Emmy_data_film.drop('Winner',axis=1,inplace=True)
```

```python
Emmy_data_film.to_csv('Emmy_Prime_only_film.csv',index=False)
```
